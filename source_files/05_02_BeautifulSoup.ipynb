```python
import requests, bs4
```

```python
res = requests.get('https://scholar.google.fr/citations?user=dXgZ_ZwAAAAJ&hl=fr')
res.text
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(res.text, 'html.parser')
print(soup.prettify())
```

```python
papiers = soup.find_all('a', class_='gsc_a_at')
papiers
```

```python
len(papiers)
```

```python
papier1 = papiers[0]
papier1.text
```

```python
citations = soup.find_all('a', class_='gsc_a_ac gs_ibl')
citations
```

```python
citation1 = citations[0]
citation1.text
```
